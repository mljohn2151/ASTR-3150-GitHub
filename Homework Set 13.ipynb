```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = 4.289e22
```

```python
G = 6.67e-11
k = 1.381e-23
m = 1.67e-27
u = .5
R = 6.956e8
M = 1.989e30
r = np.arange(0,7e8,1e8)
```

```python
def T(r):
    return ((G*m*np.pi)/(k*(1-(r/R)**2)))*(((2/3)-q)*R**2 + (8/9)*R - (2/(3*R**4))*r**6 + (q/R**2)*r**3 - (4/3)*r)
```

```python
t = T(r)
plt.plot(r,t)
plt.xlabel('Distance from the Center of the Sun (m)')
plt.ylabel('Temperature (K)')
```

```python
def P(r):
    return -G*q*np.pi*(((2/3)-q)*R**2 + (8/9)*R - (2/(3*R**4))*r**6 + (q/R**2)*r**3 - (4/3)*r)
```

```python
pressure = P(r)
plt.plot(r,pressure)
plt.xlabel('Distance from the Center of the Sun (m)')
plt.ylabel('Pressure (Pa)')
```
