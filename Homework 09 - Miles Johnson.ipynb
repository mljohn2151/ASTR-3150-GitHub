```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def boltzmanneq (T):
    return (8)/(2) * np.exp(-10.2/((8.617e-5)*(T)))

T = np.linspace(1, 10000)
y1 = boltzmanneq(T)

plt.plot(T, y1,'b')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Hydrogen in n=2 to n=1')
```

```python
def sahaeq (T):
    return (10e-20)*((2.860507881e13)*T)**(3/2)*np.exp((-13.6)/((8.617e-5)*(T)))

y2 = sahaeq(T)
plt.plot(T, y2,'r')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Ionized Hydrogen to Neutral Hydrogen')
```
