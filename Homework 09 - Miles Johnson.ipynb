```python
print("For an Ideal Hydrogen Gas in thermodynamic equilibrium, write a Python function to compute the following quantiteis at the given temperature:")
print("Fraction of neutral hydrogen in the first excited state (n=2) as comparred to the ground state (n=1)")
print("Fration of ionized hydrogen as compared to neutral hydrogen.")
print("Fraction of neutral hydrogen in the first excited state compared to the total amount of hydrogen (neutral and ionized in any excitation state).")
print("Plot all three quantities as a function of temperature in a single figure, assuming ne = 10^20 m^-3 you may need to normalize each curve to put them on the same scale). Be sure to select and justify a suitable range of temperatures.")
```

```python
%matplotlib inline
from astropy import constants
import numpy as np
import matplotlib.pyplot as plt

def boltzmanneq (T):
    return (8)/(2) * np.exp(-10.2/((8.617e-5)*(T)))

T = np.linspace(1, 10000000)
y1 = boltzmanneq(T)

plt.plot(T, y1,'b')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Hydrogen in n=2 to n=1')
```

```python
def sahaeq (T):
    return (10e-20)*((35.49262384e6)*T)**(3/2)*np.exp((-13.6)/((8.617e-5)*(T)))

y2 = sahaeq(T)
plt.plot(T, y2,'r')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Ionized Hydrogen to Neutral Hydrogen')
```

```python
y3 = boltzmanneq(T) * (1 + sahaeq(T))**-1
plt.plot(T,y3,'g')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of neutral hydrogen in the first excited state to the total amount of hydrogen')
```

```python
plt.plot(T, y1,'b')
plt.plot(T, y2,'r')
plt.plot(T,y3,'g')
plt.xlabel('Temperature (K)')
```
