```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def boltzmanneq (T,N,n):
    return (2*N*N)/(2*n*n) * np.exp(-10.2/((8.617e-5)*(T)))

T = np.linspace(1, 20000)
y1 = boltzmanneq(T,2,1)

plt.plot(T, y1,'b')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Hydrogen in n=2 to n=1')
```

```python
def sahaeq (T):
    return (T**3/2)*(1.532)*np.exp((-13.6)/((8.617e-5)*(T)))

y2 = sahaeq(T)
plt.plot(T, y2,'r')
plt.xlabel('Temperature (K)')
plt.ylabel('Ratio of Ionized Hydrogen to Neutral Hydrogen')
```

```python
plt.plot(T, y1,'b')
plt.plot(T, y2,'r')
```
