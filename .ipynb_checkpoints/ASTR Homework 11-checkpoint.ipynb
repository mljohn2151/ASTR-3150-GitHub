```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.arange(0,3*np.pi)
```

```python
def r(theta,en):
    return((1-(en**2))/(1+(en*np.cos(theta))))
```

```python
e1 = plt.subplot(111, projection='polar')
e1.plot(r(theta,0))
```

```python
e1 = plt.subplot(111, projection='polar')
e1.plot(r(theta,.5))
```
